```python
prompt = "Person's age?"
prompt += "\nEnter 'done' when you are finished: "

while True:
  age = (input(prompt))
  if age == 'done':
    break
  age = int(age)
  
  if age < 3:
    print("The person gets a free ticket!")
  elif 3 <= age <= 12:
    print("The person's ticket is $10.")
  else:
    print("The person's ticket is $15.")
    

    

```

##7-8 Deli

```python
sandwich_orders = ['chicken-sandwich','egg-sandwich',
                    'seafood-sandwich','nutella-sandwich','grilled-cheese']

finished_sandwiches = []

while sandwich_orders:
  current_order = sandwich_orders.pop()

  print(f"I made your {current_order.title()}!\n")
  finished_sandwiches.append(current_order)


print(f"The Following sandwiches have been made:\n")
for finished in finished_sandwiches:
  
  print(f"-{finished}\n")
```

##7-9 No Pastrami

```python
sandwich_orders = ['chicken-sandwich','egg-sandwich','pastrami',
                    'seafood-sandwich','pastrami',
                    'nutella-sandwich','grilled-cheese','pastrami']

finished_sandwiches = []

print("The Deli has run out of pastrami...\n\n")


while 'pastrami' in sandwich_orders:
  sandwich_orders.remove('pastrami')
  



while sandwich_orders:
  current_order = sandwich_orders.pop()

  print(f"I made your {current_order.title()}!\n")
  finished_sandwiches.append(current_order)


print(f"The Following sandwiches have been made:\n")
for finished in finished_sandwiches:
  
  print(f"{finished.title()}\n")
```

##7-10 Dream Vacation

```python

#Create an empty dictionary for the inputs to be stored.
vacationists = {}

#Polling state = True
active = True

#While-loop to create the sequence of questions and gather data.
while active:
  name = input("What is your name? \n")
  response = input("Where would you like to go someday? \n")

  #Store the user inputs into the empty dictionary.
  vacationists[name] = response

  #Find out if anyone else is going to take the poll.
  pollers = input("Is there anyone else that wants to take the poll? (yes/no) ")
  if pollers == 'no':
    active = False
    

#Polling is complete. Show the results.

print("---- POLL RESULTS ----")

for name, response in vacationists.items():
  print(f"{name.title()} wants to go '{response.title()}' someday, we hope that your wish become true!\n")

```
